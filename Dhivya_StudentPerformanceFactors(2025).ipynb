# Splitting dataset for training

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['Hours_Studied', 'Attendance', 'Parental_Involvement',
       'Access_to_Resources', 'Extracurricular_Activities', 'Previous_Scores',
       'Internet_Access', 'Tutoring_Sessions', 'Teacher_Quality',
       'Peer_Influence', 'Learning_Disabilities', 'Parental_Education_Level',
       'Distance_from_Home']]  # Features
y = df['Exam_Score']  # Target variable
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(X_train)
x_test_scaled = scaler.transform(X_test)
```

# Model buiding and Evaluation

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

## Linear Regression

```python
# Train a simple Linear Regression model
model = LinearRegression()
model.fit(x_train_scaled, y_train)

# Make predictions on the test set
ypred = model.predict(x_test_scaled)
```

```python
print(f"R2 value : {r2_score(y_test,ypred)}")
print(f"Mean Absolute Error : {mean_absolute_error(y_test,ypred)}")
print(f"Mean Squared Error : {mean_squared_error(y_test,ypred)}")
print(f"Root Mean Squared Error : {mean_squared_error(y_test,ypred)**0.5}")
```

```python
k_values = list(range(1, 21))
rmse_list = []
r2_list = []
for k in k_values:
  knn=KNeighborsRegressor(n_neighbors=k)
  knn.fit(x_train_scaled,y_train)
  y_pred=knn.predict(x_test_scaled)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

rmse_list.append(rmse)
r2_list.append(r2)
```



# Comparing different model

```python
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(max_depth=1),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "KNN Regressor": KNeighborsRegressor(n_neighbors=1)
}
```

```python
results = []

for name, model in models.items():
    model.fit(x_train_scaled, y_train)
    y_pred = model.predict(x_test_scaled)

    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    results.append({
        "Model": name,
        "MAE": round(mae, 3),
        "RMSE": round(rmse, 3),
        "R² Score": round(r2,3)
})
```

```python
results_df = pd.DataFrame(results).sort_values(by="RMSE")
print("\n📊 Regression Model Performance Comparison:\n")
print(results_df.reset_index(drop=True))
```

Linear Regression	Simple linear model assuming linear relationships.
KNN Regressor	Predicts based on average of k nearest neighbors.
Decision Tree	Tree-based model that splits features based on rules.

📊 Performance Metrics Explained

MAE (Mean Absolute Error):
Average of absolute differences between actual and predicted values. Lower is better.

RMSE (Root Mean Squared Error):
Square root of the average squared errors. Penalizes larger errors more. Lower is better.

R² Score (R-squared):
Proportion of variance in target variable explained by the model. Ranges from 0 to 1. Higher is better.
